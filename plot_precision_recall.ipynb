# This code is to generate Precision/Recall curve

Precision and recall value has been generated and stored in Precision.txt and Recall.txt. Then graph has been generated based on that.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import spline

file_precision=open("Precision_result.txt")
precision = file_precision.read().split(' ')
lenght_precision=len(precision)


precision.remove(precision[lenght_precision-1])


file_recall=open("recall_result.txt")
recall= file_recall.read().split(' ')
lenght_recall=len(recall)


recall.remove(recall[lenght_recall-1])

prec_recall=zip(precision, recall)

def f7(seq):
    seen = set()
    seen_add = seen.add
    return [x for x in seq if not (x in seen or seen_add(x))]
    
exclusive_prec_recall=f7(prec_recall)



prec_exclusive=[x[0] for x in exclusive_prec_recall]
recall_exclusive=[x[1] for x in exclusive_prec_recall]

plt.plot(prec_exclusive, recall_exclusive)
# plt.plot(x_smooth, y_smooth)
plt.ylabel('Recall')
plt.xlabel('Precision')
plt.title('Precision/Recall graph for 0001.jpg')
plt.show()


```
